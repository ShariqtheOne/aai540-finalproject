```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dataprep.eda import plot, plot_correlation

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (confusion_matrix, precision_score, recall_score,
                             roc_auc_score, ConfusionMatrixDisplay, f1_score,
                             accuracy_score, classification_report)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

import os
from google.colab import drive
drive.mount('/content/drive')
os.chdir(r'/content/drive/My Drive/AAI courses/AAI 540 MLOps/final project')
```

```python
df = pd.read_csv("data/Churn_Modelling.csv")
df
```

```python
df.drop(columns=["RowNumber", "Surname"], inplace=True)
```

We need to encode Gender and Geography before feeding them to a model

```python
gender_encoder = LabelEncoder()
gender_encoder.fit(["Female", "Male"])

df["Gender"] = gender_encoder.transform(df["Gender"])

geography_encoder = OneHotEncoder(categories=[['France', 'Spain', 'Germany']],
                                  drop="first", # to avoid the dummy variable trap
                                  sparse_output=False)

col_transformer = make_column_transformer((geography_encoder, ["Geography"]),
                                          remainder='passthrough')

df_encoded = pd.DataFrame(col_transformer.fit_transform(df),
                          columns=[col.replace("remainder__", "") for col in
                                   col_transformer.get_feature_names_out()])
```

```python
cols_to_use = ["onehotencoder__Geography_Spain",
               "onehotencoder__Geography_Germany",
               "CreditScore",
               "Age",
               "Tenure",
               "Balance",
               "NumOfProducts",
               "HasCrCard",
               "IsActiveMember",
               "EstimatedSalary",
               "Gender"]
target_variable = "Exited"

X = df_encoded[cols_to_use]
y = df[target_variable]
```

```python
for column in X.columns:
  X[column] = pd.to_numeric(X[column])
```

Let's split the data into training, testing, and validation

```python
X_train, X_test_val, y_train, y_test_val = train_test_split(
 X, y,
 test_size=0.2, random_state=1)

X_test, X_val, y_test, y_val = train_test_split(
 X_test_val, y_test_val,
 test_size=0.5, random_state=1)
```

We will rescale the data now.  

Note that we experimented with having the one-hot-encoded features and the other binary features scaled versus not scaled when feeding them to the model and we saw that for both Logistic Regression and XGBoost, it's better to not scale them.

```python
sc_cols = ["CreditScore",	"Age",	"Tenure",	"Balance",	"NumOfProducts",	"EstimatedSalary"]
sc = StandardScaler()

X_train[sc_cols] = sc.fit_transform(X_train[sc_cols])

X_test_val[sc_cols] = sc.transform(X_test_val[sc_cols])
X_val[sc_cols] = sc.transform(X_val[sc_cols])
X_test[sc_cols] = sc.transform(X_test[sc_cols])
```

As a baseline, let's build a logistic regression model to evaluate the performance.

```python
clf = LogisticRegression(random_state=1)
clf = clf.fit(X_train, y_train)

y_hat = clf.predict(X_train) # Predict the response for train dataset
y_pred = clf.predict(X_test_val) # Predict the response for test dataset

print("Train Accuracy:", accuracy_score(y_train, y_hat))
print("Test Accuracy:", accuracy_score(y_test_val, y_pred))
print("Precision:", precision_score(y_test_val, y_pred))
print("Recall:", recall_score(y_test_val, y_pred))
print("f1-score:", f1_score(y_test_val, y_pred))
cm = confusion_matrix(y_test_val, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm,
                       display_labels=["No churn", "churn"]).plot()
plt.title("Logistic Regression")
plt.show()
```

The results are pretty bad as we would expect. The recall is very low which means that we're able to find the percentage of the customers who churn as the value of the recall.

Let's experiment with another model type, XGBoost and see how the performance differs. We'll keep in mind resampling techniques to maybe try out afterwards.

```python
clf = xgb.XGBClassifier(objective="binary:logistic")
clf.fit(X= X_train, y=y_train, eval_set=[(X_train, y_train), (X_val, y_val)])
```

```python
eval_result = clf.evals_result()
training_rounds = range(len(eval_result['validation_0']['logloss']))

plt.scatter(x=training_rounds,y=eval_result['validation_0']['logloss'],label='Training Error')
plt.scatter(x=training_rounds,y=eval_result['validation_1']['logloss'],label='Validation Error')
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('log-loss')
plt.title('Training Vs Validation Error')
plt.legend()
plt.show()
```

We clearly see that the model is overfitting. Let's try to curb this.

Let's experiment with eta (the learning rate) for XGBoost to see if we can curb the overfitting

```python
learning_rate = [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1]

for curr_eta in learning_rate:
  clf = xgb.XGBClassifier(objective="binary:logistic", learning_rate=curr_eta)
  clf.fit(X= X_train, y=y_train, eval_set=[(X_train, y_train), (X_val, y_val)], verbose=0)

  eval_result = clf.evals_result()
  training_rounds = range(len(eval_result['validation_0']['logloss']))

  plt.scatter(x=training_rounds,y=eval_result['validation_0']['logloss'],label='Training Error')
  plt.scatter(x=training_rounds,y=eval_result['validation_1']['logloss'],label='Validation Error')
  plt.grid(True)
  plt.xlabel('Iteration')
  plt.ylabel('log-loss')
  plt.title(f'eta:{curr_eta} - Training Vs Validation Error')
  plt.legend()
  plt.show()
```

We see that the best value of eta is between 0.01 and 0.1.  

To be more scientific and precise when comparing models, we can use cross-validation to be more certain of the findings. It would also be helpful to also use stratified k folds cross valdiation to preserve the ratios of churn/no church customers while also gauging the robustness of the performance against different cuts.  

We will also use `scale_pos_weight` to increase the penality for wrong predictions in the positive class (customers who churn) which we found to help with overfitting. We will reduce `max_depth` from the default 6 to 3 to also help with overfitting.

```python
val_cnts = y_train.value_counts()
scale_pos_weight = val_cnts[0] / val_cnts[1]

clf = xgb.XGBClassifier(objective="binary:logistic",
                        learning_rate=0.07,
                        scale_pos_weight=scale_pos_weight,
                        max_depth=3)
clf.fit(X= X_train, y=y_train,
        eval_set=[(X_train, y_train), (X_val, y_val)],
        verbose=0)

y_hat = clf.predict(X_train) # Predict the response for train dataset
y_pred = clf.predict(X_test_val) # Predict the response for test dataset

print("Train Accuracy:", accuracy_score(y_train, y_hat))
print("Test Accuracy:", accuracy_score(y_test_val, y_pred))
print("Precision:", precision_score(y_test_val, y_pred))
print("Recall:", recall_score(y_test_val, y_pred))
print("f1-score:", f1_score(y_test_val, y_pred))
cm = confusion_matrix(y_test_val, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm,
                       display_labels=["No churn", "churn"]).plot()
plt.title("XGBoost")
plt.show()
```

We can see that XGBoost performs a lot better than Logistic Regression. We can certainly perform more hyperparameter tuning to see if we can get better results.

Let's examine feature importances

```python
feature_importance = clf.feature_importances_
sorted_idx = np.argsort(feature_importance)
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(X.columns)[sorted_idx])
plt.title('Feature Importance')
plt.show()
```

As we can see in the plot above, age is the biggest factor in predicting churn, followed by NumOfProducts. This is to be expected given that we saw them as having the highest correlation with the target variable in the analysis notebook.
